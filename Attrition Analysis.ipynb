```python
numeric_cols=df.select_dtypes(include=np.number).columns
```

```python
extra_info=["DailyRate","StandardHours","Education","EmployeeCount","EmployeeNumber","HourlyRate","JobInvolvement","MonthlyRate","NumCompaniesWorked","StamdardHours","StockPotionLevel","TrainingTimesLastYear","YearsInCurrentRole"]
for col in numeric_cols:
    if col not in extra_info :
        sns.set_style("whitegrid")
        df[col].hist(bins=20,figsize=(4,5),edgecolor='black')
        plt.suptitle(f"Distrribution of {col}")
        plt.show()              
```

```python
cat_cols=df.select_dtypes(include=object).columns
```

```python
for col in df[cat_cols]:
    if col not in ['Attrition']:
        sns.set_style("whitegrid")
        plt.figure(figsize=(5,4))
        ax= sns.countplot(x=col,data=df[cat_cols],edgecolor="black")
        for container in ax.containers:
            ax.bar_label(container)
        plt.title(f'plot of {col}')
        plt.xticks(rotation=50)
        plt.show()
```

numerical relationship with Attrition to find the area having most attrrition

```python
for col in numeric_cols:
    if col not in extra_info:
        sns.set_style('whitegrid')
        sns.boxplot(x='Attrition',y=col,data=df)
        plt.title(f'Attrition vs {col}')
        plt.show()
```

```python
for col in df[cat_cols]:
    if col not in ['Attrition']:
        sns.set_style('whitegrid')
        plt.figure(figsize=(4,5))
        sns.countplot(x='Attrition',hue=col,data=df,edgecolor='black')
        plt.title(f'Attriton vs {col}')
        plt.show()
```

extracting insights from combination of variable to find deep place where problem occurs

```python
sns.scatterplot(x='TotalWorkingYears',y='MonthlyIncome',hue='Attrition',data=df)
```

```python
bins = [18, 25, 35, 45, 60, df['Age'].max()]
labels = ['18-24', '26-35', '36-45', '46+']

df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=bins,
    labels=labels,
    right=False,
    duplicates='drop'  
)
```

```python
pd.crosstab([df['AgeGroup'], df['JobLevel'],df['Gender']], df['Attrition']).plot(kind='bar')
```

```python
df['ManagerTenureGroup'] = pd.cut(df['YearsWithCurrManager'], bins=[0, 2, 5, 10, 20], labels=['0–2', '3–5', '6–10', '10+'])
pd.crosstab([df['ManagerTenureGroup'], df['RelationshipSatisfaction']], df['Attrition']).plot(kind='line', stacked=True)
```
